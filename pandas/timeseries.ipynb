```python
from pandas.tseries.holiday import USFederalHolidayCalendar    ##geting hoilday calendar for us
from pandas.tseries.offsets import CustomBusinessDay

usb = CustomBusinessDay(calendar = USFederalHolidayCalendar())
usb
```

```python
rng = pd.date_range(start = '7/01/2017', end = '8/8/2017' , freq = usb)
rng
```

```python
df.set_index(rng , inplace=True)
df
```

```python
from pandas.tseries.holiday import AbstractHolidayCalendar , nearest_workday , Holiday

class PakistanHolidayCalendar(AbstractHolidayCalendar):
    rules=[
        Holiday('Independace Day' , month=8 , day = 14 , observance = nearest_workday),  ## If the holliday to on staurday or sunday it will be observed on previous or the next day
        Holiday('Kashmir Day' , month=2 , day = 5 ),
        Holiday('Lahore Resolutin Day' , month=3 , day = 23 ),
        Holiday('Labour Day' , month=5 , day = 1 ),
    ]
    
pyc = CustomBusinessDay(calendar = PakistanHolidayCalendar())
pyc
```

```python
pd.date_range(start="8/4/2020" , end="8/29/2020" , freq = pyc)
```

```python
b = CustomBusinessDay( weekmask='Sun Mon Tue Wed Thu' , holidays=["2020-08-04"]) ## setting weekend as Friday and Staurday
pd.date_range(start="8/4/2020" , end="8/29/2020" , freq = b)
```
