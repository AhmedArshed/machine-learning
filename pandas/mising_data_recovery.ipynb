```python
import pandas as pd
```

```python
df = pd.read_csv("weather.csv" , parse_dates=["date"])
```

```python
type(df.date[0])
```

```python
df.set_index('date' , inplace = True)
df
```

```python
new_df = df.fillna(0)
new_df
```

```python
new_df = df.fillna({
    'param': "not Precipitation",
    'siteid' : "not decided"
})
new_df
```

```python
new_df = df.fillna(method="ffill")  ##forward fill values
new_df
```

```python
new_df = df.fillna(method="bfill")   ##backword fill values
new_df
```

```python
new_df = df.fillna(method="pad")
new_df
```

```python
new_df = df.fillna(method="ffill" , axis = "columns")   ##copy values from prvious cell of coloume
new_df
```

```python
new_df = df.fillna(method="ffill" , limit = 1)   ##copy values only onces
new_df
```

```python
new_df = df.interpolate()  ##interpolate is use for numerical valued data
new_df
```

```python
new_df = df.interpolate(method = "time")  ## time will get a value nearest to that time
new_df
```

```python
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how="all")  ##drop the cells having all nan values
new_df
```

```python
new_df = df.dropna(thresh=2)  ##having atlest 2 non NaN 
new_df
```
