```python
import pandas as pd
df = pd.read_csv("weather.csv" , parse_dates = ["date"])
df
```

```python
df.pivot(index="city" , columns="date")
```

```python
df.pivot(index="city" , columns="date" , values = "event")
```

```python
df.pivot(index="event" , columns="city" , values="tempreature")
```
